Question 1. What is Anomaly Detection? Explain its types with examples.

Anomaly Detection is the process of identifying unusual patterns or observations in data that do not follow the expected normal behavior. These unusual observations are called anomalies or outliers. Detecting anomalies is important because they may indicate critical events such as fraud, system failure, network intrusion, or sudden spikes in demand.

For example, in banking transactions, if a customer usually spends around 1000 to 2000 rupees daily and suddenly makes a transaction of 1 lakh rupees, that transaction is considered an anomaly and may indicate fraud.

Types of Anomalies

Point Anomaly
A point anomaly occurs when a single data point is significantly different from the rest of the data.

Example
Daily electricity consumption values for a household:
10 units, 12 units, 11 units, 50 units, 12 units
Here, 50 units is a point anomaly because it is very different from the surrounding values.

Use case
Fraud detection, faulty sensor readings, sudden spike in server load.

Contextual Anomaly
A contextual anomaly occurs when a data point is considered abnormal in a specific context such as time, season, or location.

Example
Temperature of 10 degrees Celsius
This is normal in winter but abnormal in summer.
So, the same value can be normal or anomalous depending on the context.

Use case
Seasonal sales analysis, energy consumption monitoring, weather pattern analysis.

Collective Anomaly
A collective anomaly occurs when a group of data points together behaves abnormally even if individual points may seem normal.

Example
Normal heart rate pattern suddenly changes continuously for several readings. Each reading alone may look acceptable, but the overall pattern is abnormal.

Use case
Network intrusion detection, equipment failure detection, medical monitoring systems.

Question 2. Compare Isolation Forest, DBSCAN, and Local Outlier Factor in terms of approach and suitable use cases.

Isolation Forest
Isolation Forest is an ensemble tree based algorithm. It works by randomly selecting a feature and splitting the data. Anomalies are easier to isolate because they are far from normal data points, so they require fewer splits to be separated. The shorter the path length in the tree, the higher the chance that the point is an anomaly.

Approach
It isolates anomalies instead of profiling normal points.

Suitable use cases
Large high dimensional datasets, fraud detection, network security monitoring, large scale log analysis.

Example
Detecting abnormal taxi fares where extremely high or extremely low fares are isolated quickly by the model.

DBSCAN
DBSCAN is a density based clustering algorithm. It groups data points based on density. Points in high density regions form clusters, while points in low density regions are considered noise or anomalies.

Approach
It identifies anomalies as points that do not belong to any dense cluster.

Suitable use cases
Geospatial data, customer segmentation, anomaly detection in location based services.

Example
In GPS location data, most taxi pickups occur in busy areas forming dense clusters. Pickups occurring in isolated remote areas are treated as anomalies.

Local Outlier Factor (LOF)
Local Outlier Factor measures the local density of a data point and compares it with the density of its neighbors. If the density of a point is much lower than its neighbors, it is considered an anomaly.

Approach
It detects anomalies based on local density deviation.

Suitable use cases
Detecting local anomalies in datasets where normal behavior varies across regions.

Example
In credit card data, spending of 5000 rupees may be normal for high income customers but anomalous for low income customers. LOF captures this local variation effectively.

Question 3. What are the key components of a Time Series? Explain each with one example.

A time series is a sequence of data points collected over time at regular intervals. It has four main components: trend, seasonality, cyclic patterns, and residual noise.

Trend
Trend represents the long term movement or direction of the data over time. It shows whether the data is increasing, decreasing, or remaining stable.

Example
Yearly growth in company revenue increasing every year shows an upward trend.

Seasonality
Seasonality refers to repeating patterns at fixed intervals such as daily, monthly, or yearly cycles.

Example
Ice cream sales increase every summer and decrease in winter. This repeating yearly pattern is seasonality.

Cyclic Component
Cyclic patterns are long term fluctuations that do not occur at fixed intervals. They are usually influenced by economic conditions or business cycles.

Example
Economic boom and recession cycles affecting stock market performance over several years.

Residual or Noise
Residual represents random variations that cannot be explained by trend, seasonality, or cyclic patterns.

Example
Sudden unexpected spike in website traffic due to a viral social media post.

Question 4. Define Stationarity in time series. How can you test and transform a non stationary series into a stationary one?

Stationarity means that the statistical properties of a time series such as mean, variance, and autocorrelation remain constant over time. Many forecasting models like ARIMA require the time series to be stationary.

In a stationary series
The mean remains constant over time
The variance remains constant
There is no strong trend or seasonal pattern

Testing Stationarity

Visual Inspection
Plot the time series. If the mean and variance change over time, the series is non stationary.

Augmented Dickey Fuller Test
This statistical test checks whether a time series is stationary.
If p value is less than 0.05, the series is considered stationary.

Transforming Non Stationary to Stationary

Differencing
Subtract the previous observation from the current observation. This removes trend.

Example
If values are 100, 120, 140, 160
Differenced series becomes 20, 20, 20 which is stationary.

Log Transformation
Used to stabilize variance when variance increases over time.

Seasonal Differencing
Subtract value from previous season. Useful for seasonal patterns.

Question 5. Differentiate between AR, MA, ARIMA, SARIMA, and SARIMAX models.

AR Model
Auto Regressive model predicts the current value using past values of the same series.

Structure
Yt depends on Yt minus 1, Yt minus 2, and so on.

Application
Stock price prediction where current price depends on previous prices.

MA Model
Moving Average model uses past forecast errors to predict current values.

Structure
Yt depends on previous error terms.

Application
Noise smoothing and short term forecasting.

ARIMA Model
ARIMA combines AR and MA along with differencing to make the series stationary. It is used for non seasonal time series forecasting.

Structure
AR part plus differencing plus MA part.

Application
Forecasting sales data without seasonal patterns.

SARIMA Model
SARIMA extends ARIMA by including seasonal components. It captures both trend and seasonality.

Application
Monthly airline passenger data where seasonal yearly patterns exist.

SARIMAX Model
SARIMAX is SARIMA with external variables. It includes exogenous features such as weather, promotions, or holidays.

Application
Energy demand forecasting using temperature, holidays, and time as additional inputs.

Question 6. Load a time series dataset such as AirPassengers, plot the original series, and decompose it into trend, seasonality, and residual components.

Explanation
The AirPassengers dataset contains monthly totals of airline passengers from 1949 to 1960. Decomposition helps us understand three components
Trend shows long term increase or decrease
Seasonality shows repeating patterns at fixed intervals
Residual shows random noise

Python Code

import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.datasets import get_rdataset
from statsmodels.tsa.seasonal import seasonal_decompose

Load AirPassengers dataset

data = get_rdataset('AirPassengers').data
data['time'] = pd.date_range(start='1949-01', periods=len(data), freq='M')
data.set_index('time', inplace=True)

Plot original series

plt.figure()
plt.plot(data['value'])
plt.title('AirPassengers Time Series')
plt.xlabel('Year')
plt.ylabel('Passengers')
plt.show()

Decompose time series

decomposition = seasonal_decompose(data['value'], model='multiplicative')

Plot components

decomposition.plot()
plt.show()

Interpretation
The trend shows steady growth in airline passengers over the years.
The seasonal component shows repeated yearly peaks and drops.
The residual shows irregular fluctuations not explained by trend or seasonality.

Question 7. Apply Isolation Forest on a numerical dataset to detect anomalies and visualize them using a scatter plot.

Explanation
Isolation Forest isolates anomalies using random trees. Points that are isolated quickly are considered anomalies.

Example Dataset
We will simulate taxi fare data with distance and fare amount.

Python Code

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

Create synthetic taxi fare dataset

np.random.seed(42)
distance = np.random.normal(5, 2, 200)
fare = distance * 10 + np.random.normal(0, 5, 200)

Add anomalies

distance_outliers = np.array([20, 22, 25])
fare_outliers = np.array([200, 220, 250])

distance = np.concatenate([distance, distance_outliers])
fare = np.concatenate([fare, fare_outliers])

df = pd.DataFrame({'distance': distance, 'fare': fare})

Apply Isolation Forest

model = IsolationForest(contamination=0.05)
df['anomaly'] = model.fit_predict(df)

Plot results

plt.figure()
plt.scatter(df['distance'], df['fare'], c=df['anomaly'])
plt.title('Isolation Forest Anomaly Detection')
plt.xlabel('Distance')
plt.ylabel('Fare')
plt.show()

Interpretation
Normal points are grouped together while extreme fare and distance values are marked as anomalies.

Question 8. Train a SARIMA model on the monthly airline passengers dataset and forecast the next 12 months.

Explanation
SARIMA handles seasonality in time series. AirPassengers data has yearly seasonal patterns, so SARIMA is suitable.

Python Code

import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.datasets import get_rdataset
from statsmodels.tsa.statespace.sarimax import SARIMAX

Load dataset

data = get_rdataset('AirPassengers').data
data['time'] = pd.date_range(start='1949-01', periods=len(data), freq='M')
data.set_index('time', inplace=True)

Train SARIMA model

model = SARIMAX(data['value'], order=(1,1,1), seasonal_order=(1,1,1,12))
results = model.fit()

Forecast next 12 months

forecast = results.forecast(steps=12)

Plot

plt.figure()
plt.plot(data['value'], label='Original')
plt.plot(forecast, label='Forecast')
plt.legend()
plt.title('SARIMA Forecast for AirPassengers')
plt.show()

Interpretation
The forecasted line continues the upward trend and repeating seasonal peaks similar to historical data.

Question 9. Apply Local Outlier Factor on any numerical dataset to detect anomalies and visualize them.

Explanation
Local Outlier Factor compares density of each point with its neighbors. Points with much lower density are considered anomalies.

Python Code

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import LocalOutlierFactor

Create synthetic dataset

np.random.seed(0)
X = 0.3 * np.random.randn(200, 2)

Add outliers

outliers = np.random.uniform(low=-4, high=4, size=(20, 2))
X = np.concatenate([X, outliers])

Apply LOF

lof = LocalOutlierFactor(n_neighbors=20, contamination=0.1)
y_pred = lof.fit_predict(X)

Plot

plt.figure()
plt.scatter(X[:, 0], X[:, 1], c=y_pred)
plt.title('Local Outlier Factor Anomaly Detection')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()

Interpretation
Dense central cluster represents normal data points. Scattered points far away are detected as anomalies.

Question 10. Real time data science workflow for power grid monitoring company.

Problem
We receive energy consumption data every 15 minutes along with timestamp, region, weather, and energy usage. We must forecast demand and detect abnormal spikes or drops.

Step 1. Detect anomalies in streaming data
We can use Isolation Forest for real time detection because it is fast and works well with large streaming data. LOF can also be used when local consumption patterns vary by region. DBSCAN can be used when data forms clear density clusters but is less suitable for real time streaming.

Workflow
Collect streaming data every 15 minutes
Extract features such as rolling mean, rolling standard deviation
Feed data into Isolation Forest model
Flag sudden spikes or drops as anomalies
Send alert to monitoring dashboard

Step 2. Choose time series model for short term forecasting
SARIMAX is the best choice because it can handle seasonality and also include external features like weather and region.

Why not ARIMA
ARIMA does not handle seasonality or external variables effectively.

Step 3. Validate and monitor performance
Use metrics such as Mean Absolute Error and Root Mean Square Error to compare predicted and actual energy usage.
Perform rolling forecast validation where model is retrained periodically.
Monitor model drift by checking if error increases over time.

Step 4. Business impact
Early detection of abnormal spikes prevents power outages and equipment overload.
Accurate demand forecasting helps optimize power generation and distribution.
Helps reduce operational cost and ensures reliable electricity supply.
