4.  How does loss value help in determining whether the model is good or not?
Ans: The loss value tells us how far the model‚Äôs predictions are from the correct answers.

Think of it like a score that measures the model‚Äôs mistakes.

‚úî What low loss means

‚úî What high loss means

 6.How do we handle categorical variables in Machine Learning? What are the common techniques?
 Ans:How do we handle categorical variables in Machine Learning?

Machine Learning models work with numbers, not text.
So categorical variables (like "red", "blue", "green") must be converted into numbers.

This process is called encoding.

Common Techniques to Handle Categorical Variables
1. Label Encoding

Converts each category into a unique number

Example:
red ‚Üí 0
blue ‚Üí 1
green ‚Üí 2

‚úî Useful for ordinal categories (categories with order)
‚úñ Not ideal for categories without order, because numbers may introduce false ranking

2. One-Hot Encoding

Creates separate columns for each category

Values are 1 or 0

Example for colors:
color_red, color_blue, color_green

‚úî Best for categories with no order
‚úî Very commonly used
‚úñ Increases number of columns if categories are many

3. Ordinal Encoding

Assigns numbers based on order

Useful for ordered categories
Example:
small ‚Üí 1
medium ‚Üí 2
large ‚Üí 3

‚úî Works when category order matters

4. Target Encoding

Replaces each category with the average of the target value for that category
‚úî Useful for high-cardinality features (many unique values)
‚úñ Risk of overfitting if not used carefully

5. Binary Encoding / Hash Encoding

Converts categories into binary digits
‚úî Useful when there are many categories
‚úî Reduces dimensionality
‚úñ Harder to interpret

Summary Table
Technique	When to Use
Label Encoding	Ordered categories (low < medium < high)
One-Hot Encoding	Unordered categories (red, blue, green)
Ordinal Encoding	Categories with clear order
Target Encoding	High-cardinality categories

9. What is a Test set?
Ans: What is a Test Set?

A test set is the portion of the dataset that is used to evaluate the performance of a Machine Learning model after it has been trained.

The model does not see or learn from this data during training.

Why do we use a Test Set?

Because we want to check:

How well the model performs on new, unseen data

Whether the model is generalizing properly

Whether it is overfitting (memorizing the training data)

Simple Example:

Imagine you are teaching a student:

Training set = questions you use to teach

Test set = new questions you give during an exam

If the student does well in the exam, it means they understood the concept ‚Äî not just memorized.

Similarly, a test set checks if the model has actually learned.

Summary

Test set = Data used to evaluate the model

Model never sees this data during training

Helps measure real-world performance

Step 1: Understand the Problem

Identify the goal (classification, regression, clustering)

Understand what prediction is needed

Load the dataset

Check missing values

Understand correlations

Handle missing values

Handle categorical variables (encoding)

Scale or normalize numeric data

Use metrics such as accuracy, precision, recall, RMSE, etc.

Compare performance across models

Step 9: Improve the Model

Hyperparameter tuning

Try different models

Add/remove features

Step 10: Deploy the Model

Convert the model into a usable form (API, web app, etc.)

16. What is an Optimizer? What are different types of optimizers? Explain each with an example.

An optimizer is a method used during model training to reduce the loss value and improve accuracy.
It adjusts the model parameters step by step so the model learns correctly.

Here are the common optimizers:

a. Gradient Descent

It moves the parameters slowly in the direction that reduces loss.
Example:
Imagine you are climbing down a hill in the dark by taking small steps until you reach the bottom.
The bottom represents the lowest loss.

b. Stochastic Gradient Descent

Instead of using the whole dataset at once, it updates parameters using one sample at a time.
This makes training faster.

c. Mini Batch Gradient Descent

It uses small batches of the data at each step.
This is the most commonly used method because it is both fast and stable.

d. Adam Optimizer

Adam is an advanced optimizer that adjusts the learning rate automatically during training.
It is faster and often gives better results.

Example:
Deep learning models like image recognition usually use Adam.
