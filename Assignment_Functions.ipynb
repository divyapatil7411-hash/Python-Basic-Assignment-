# 1. What is the difference between a function and a method in Python?
Ans: A function and a method in Python are very similar (both are callable blocks of code), but the main difference lies in how they are defined and used:

**Function**

Defined using def or lambda.

Can exist on its own, outside of any class.

Called directly using its name.

Example:

def greet(name):
    return f"Hello, {name}!"

print(greet("Divya"))


**Method**

A function that is defined inside a class.

Always associated with an object or class.

The first parameter of an instance method is usually self, which refers to the object itself.

Example:

class Greeter:
    def greet(self, name):
        return f"Hello, {name}!"

obj = Greeter()
print(obj.greet("Divya"))

#2. Explain the concept of function arguments and parameters in Python.
Ans:parameters and arguments are related but not the same:

**Parameters** â†’ the variables listed in the function definition.

**Arguments** â†’ the actual values you pass when calling the function.

**Example**:

**Positional arguments** - Values are passed in order.

**Keyword arguments** - Specify by name.

    print(f"My name is {name}, and I am {age} years old.")

**Default arguments** - Provide default values.

def greet(name="Divya"):
    print(f"Hello, {name}!")

greet()          # uses default "Divya"
greet("Ramesh")  # overrides default


**Variable-length arguments -**

*args for multiple positional arguments.

**kwargs for multiple keyword arguments.

def show_friends(*args):
    for friend in args:
        print(friend)

show_friends("Divya", "Ramesh", "Amit")

#5. What are iterators in Python and how do they differ from iterables?
Ans:**Iterable**

An iterable is any Python object that can return its elements one at a time.

Examples: lists, tuples, strings, dictionaries, sets.

It has the __iter__() method.

But an iterable itself does not remember its position while looping.

Example (iterable):

my_list = ["Divya", "Ramesh", "Amit"]   # list is iterable

for name in my_list:
    print(name)

**Iterator**

An iterator is an object that represents a stream of data.

It has both __iter__() and __next__() methods.

It remembers its current position and gives the next item when asked.

You can create an iterator from any iterable using iter().

Example (iterator):

my_list = ["Divya", "Ramesh", "Amit"]
it = iter(my_list)      # creating iterator

print(next(it))   # Divya
print(next(it))   # Ramesh
print(next(it))   # Amit
 print(next(it)) # Error: StopIteration (no more elements)

#6. Explain the concept of generators in Python and how they are defined.
Ans:What is a Generator?

A generator in Python is a special type of iterator that lets you generate values on the fly instead of storing them all in memory at once.

Generators are memory-efficient.

They are created using functions with yield instead of return.

Each time yield is called, the function pauses and can be resumed later.

ðŸ”¹ Defining a Generator

Generators can be defined in two main ways:

1.**Generator Function** (uses yield)

def greet_generator():

  yield "Hello Divya"

  yield "How are you?"

  yield "Goodbye Divya"

gen = greet_generator()

print(next(gen))  # Hello Divya

print(next(gen))  # How are you?

print(next(gen))  # Goodbye Divya


Here:

yield returns a value without ending the function.

The function remembers its state between calls.

2.**Generator Expression** (like list comprehension but with ())

squares = (x*x for x in range(1, 6))

for val in squares:

  print(val)


This will print squares 1, 4, 9, 16, 25 â€” but values are generated one at a time, not stored like in a list.

Difference Between Generators and Iterators

Iterator â†’ Any object with __iter__() and __next__().

Generator â†’ A shortcut to create an iterator using yield.

#7. What are the advantages of using generators over regular functions?
Ans:Advantages of Using Generators Over Regular Functions

1.**Memory Efficiency**

Regular functions that return lists store all results in memory.

Generators yield one item at a time, so they use very little memory.

def all_names():
    return ["Divya", "Ramesh", "Amit"]  # entire list in memory

def name_generator():

    yield "Divya"

    yield "Ramesh"

    yield "Amit"  # one at a time

2.**Lazy Evaluation** (On-Demand Computation)

Values are produced only when requested with next() or a loop.

Saves time when you don't need all results at once.

gen = (x*x for x in range(1_000_000))

print(next(gen))  # calculates only 1*1

3.**Support for Infinite Sequences**

Regular functions cannot return infinite lists (memory issue).

Generators can produce infinite streams.

def infinite_numbers():
  n = 1

  while True:

  yield n

  n += 1

4.**Better Performance**

Generators avoid creating large temporary data structures.

They run faster for big datasets because they generate values on the fly.

5.**Cleaner, Simpler Code**

No need to manage intermediate lists or counters.

yield makes code easier to read.

#9. Explain the purpose and usage of the 'map()' function in Python.

#10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

Ans:**1.map()**

Purpose: Applies a function to every item of an iterable and returns a map object with the results.

Use Case: Transforming data element-wise.

Example:

numbers = [1, 2, 3, 4]

squares = list(map(lambda x: x*x, numbers))

print(squares)  # [1, 4, 9, 16]

2.**filter()**

Purpose: Filters elements from an iterable based on a condition and returns a filter object containing only items where the function returns True.

Use Case: Selecting a subset of data.

Example:

numbers = [1, 2, 3, 4, 5]

even_numbers = list(filter(lambda x: x % 2 == 0, numbers))

print(even_numbers)  # [2, 4]

3.**reduce()** (from functools module)

Purpose: Applies a function cumulatively to items of an iterable, reducing it to a single value.

Use Case: Combining or aggregating data.

Example:

from functools import reduce

numbers = [1, 2, 3, 4]

sum_numbers = reduce(lambda x, y: x + y, numbers)

print(sum_numbers)  # 10

```python
#4. Write a Python function that checks if a given number is prime or not from 1 to 200.
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True

primes = [x for x in range(1, 201) if is_prime(x)]
print(primes)
```

```python
#5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
class Fibonacci:
    def __init__(self, n):
        self.n = n
        self.a, self.b = 0, 1
        self.count = 0
    def __iter__(self):
        return self
    def __next__(self):
        if self.count >= self.n:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return 0
        elif self.count == 1:
            self.count += 1
            return 1
        else:
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return self.b

fib = Fibonacci(10)
print(list(fib))
```

```python
#6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
def powers_of_two(n):
    for i in range(n+1):
        yield 2**i

for val in powers_of_two(5):
    print(val)
```

```python
# 7. Implement a generator function that reads a file line by line and yields each line as a string.
def read_file_line_by_line(filename):
    with open(filename, 'r') as f:
        for line in f:
            yield line.strip()
```
