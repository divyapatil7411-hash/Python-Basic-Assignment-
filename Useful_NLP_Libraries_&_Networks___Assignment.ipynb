# Question 1: Compare and contrast NLTK and spaCy in terms of features, ease of use, and performance.

**Answer:**

**NLTK (Natural Language Toolkit)** and **spaCy** are two popular Python libraries used for Natural Language Processing (NLP). Although both are designed for text processing, they differ in their features, ease of use, and performance.

| Feature                  | NLTK                                              | spaCy                                                   |
| ------------------------ | ------------------------------------------------- | ------------------------------------------------------- |
| Purpose                  | Mainly for learning, research, and education      | Designed for industrial and real-world NLP applications |
| Ease of Use              | Beginner-friendly with many educational resources | Easy to use with a simple and efficient API             |

### Similarities

* Both are open-source Python libraries.
* Both support tokenization, part-of-speech tagging, and named entity recognition.
* Both can be used for text preprocessing and analysis.

### Differences

* **NLTK** provides many linguistic resources and is ideal for learning NLP concepts.
* **spaCy** is optimized for speed and is commonly used in industry.
* spaCy provides better performance when processing large amounts of text.
* NLTK includes many corpora and datasets for research purposes.

### Conclusion

NLTK is best suited for beginners, students, and researchers who want to learn NLP concepts. spaCy is better suited for developers and data scientists building fast, scalable, and production-ready NLP applications.

# Question 2: What is TextBlob and how does it simplify common NLP tasks like sentiment analysis and translation?

**Answer:**

**TextBlob** is a simple and beginner-friendly Python library built on top of **NLTK** and **Pattern**. It provides an easy interface for performing common Natural Language Processing (NLP) tasks without requiring complex code.

### Features of TextBlob

* Tokenization
* Part-of-Speech (POS) Tagging
* Noun Phrase Extraction
* Sentiment Analysis
* Spell Checking
* Language Translation
* Text Classification

### How TextBlob Simplifies NLP Tasks

#### 1. Sentiment Analysis

TextBlob automatically determines the sentiment of a piece of text by calculating:

* **Polarity:** A value between -1 (negative) and +1 (positive).
* **Subjectivity:** A value between 0 (objective) and 1 (subjective).

**Example:**

```python
from textblob import TextBlob

text = TextBlob("The product is amazing and easy to use.")
print(text.sentiment)
```

**Sample Output:**

```text
Sentiment(polarity=0.6, subjectivity=0.9)
```

This indicates that the review expresses a positive opinion.

#### 2. Translation

TextBlob can translate text from one language to another using supported translation services.

**Example:**

```python
from textblob import TextBlob

text = TextBlob("Hello, how are you?")
translated = text.translate(from_lang='en', to='fr')
print(translated)
```

**Sample Output:**

```text
Bonjour, comment allez-vous ?
```

### Advantages of TextBlob

* Easy to learn and use.
* Requires very little code.
* Ideal for beginners and small NLP projects.
* Supports several common NLP tasks with simple methods.

### Limitations

* Slower than spaCy for large datasets.
* Not suitable for highly complex or production-scale NLP applications.
* Translation features may depend on external services and internet connectivity.

### Conclusion

TextBlob is an excellent library for beginners because it simplifies many common NLP tasks, including sentiment analysis, translation, tokenization, and part-of-speech tagging. It is well suited for educational purposes, small projects, and rapid prototyping.

# Question 4: Describe the architecture and functioning of a Recurrent Neural Network (RNN).

**Answer:**

A **Recurrent Neural Network (RNN)** is a type of artificial neural network designed to process sequential data such as text, speech, and time-series information. Unlike traditional neural networks, an RNN has connections that allow information from previous inputs to influence the processing of current inputs.

### Architecture of an RNN

An RNN consists of:

* **Input Layer:** Receives one word or data element at a time.
* **Hidden Layer:** Stores information from previous time steps using a hidden state (memory).
* **Output Layer:** Produces the prediction or output for each input.

The hidden state is passed from one time step to the next, allowing the network to remember previous information.

### Working of an RNN

1. The first input word is fed into the network.
2. The hidden layer processes the input and stores information.
3. The hidden state is passed to the next time step.
4. The next input is processed together with the previous hidden state.
5. This process continues until all inputs in the sequence have been processed.
6. The final output is generated based on the learned sequence information.

### Applications of RNN

* Language Translation
* Speech Recognition
* Text Generation
* Sentiment Analysis
* Chatbots
* Predictive Text

### Advantages

* Can process sequential data.
* Remembers previous information using hidden states.
* Suitable for language and speech processing tasks.

### Limitations

* Suffers from the **vanishing gradient problem** when processing long sequences.
* Has difficulty remembering long-term dependencies.
* Slower to train compared to newer models.

### Conclusion

RNNs are effective for processing sequential data because they retain information from previous inputs. However, for long sequences, models such as **LSTM** and **GRU** are generally preferred because they overcome many of the limitations of standard RNNs.

# Question 5: What is the key difference between LSTM and GRU networks in NLP applications?

**Answer:**

**LSTM (Long Short-Term Memory)** and **GRU (Gated Recurrent Unit)** are advanced versions of Recurrent Neural Networks (RNNs). Both are designed to solve the vanishing gradient problem and improve the learning of long-term dependencies in sequential data.

### Comparison of LSTM and GRU

| Feature         | LSTM                                | GRU                                     |
| --------------- | ----------------------------------- | --------------------------------------- |
| Full Form       | Long Short-Term Memory              | Gated Recurrent Unit                    |
| Number of Gates | Three gates (Input, Forget, Output) | Two gates (Reset and Update)            |
| Memory Cell     | Has a separate memory cell          | No separate memory cell                 |
| Complexity      | More complex                        | Simpler                                 |
| Training Speed  | Slower                              | Faster                                  |
| Parameters      | More parameters                     | Fewer parameters                        |
| Performance     | Better for very long sequences      | Faster and efficient for many NLP tasks |

### Key Differences

* LSTM contains **three gates** and a dedicated memory cell, allowing it to retain long-term information more effectively.
* GRU combines some of these functions into **two gates**, making it simpler and computationally efficient.
* LSTM generally performs better on complex tasks involving long sequences, while GRU often trains faster and requires less memory.

### Applications

**LSTM**

* Machine Translation
* Speech Recognition
* Text Generation
* Language Modeling

**GRU**

* Chatbots
* Sentiment Analysis
* Text Classification
* Predictive Text

### Conclusion

Both LSTM and GRU are powerful models for NLP applications. LSTM is preferred for tasks requiring long-term memory and complex sequence learning, while GRU is often chosen when faster training and lower computational cost are important.

# Question 8: Implement a basic LSTM model in Keras for a text classification task

**Answer:**

### Python Code

```python id="lstm_q8_keras"
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense

# Dataset
texts = [
    "I love this project",
    "This is an amazing experience",
    "I hate waiting in line",
    "This is the worst service",
    "Absolutely fantastic!"
]

labels = np.array([1, 1, 0, 0, 1])

# Step 1: Tokenization
tokenizer = Tokenizer()
tokenizer.fit_on_texts(texts)

sequences = tokenizer.texts_to_sequences(texts)

print("Word Index:")
print(tokenizer.word_index)

print("\nSequences:")
print(sequences)

# Step 2: Padding sequences
max_len = max(len(seq) for seq in sequences)
padded_sequences = pad_sequences(sequences, maxlen=max_len, padding='post')

print("\nPadded Sequences:")
print(padded_sequences)

# Step 3: Build LSTM model
vocab_size = len(tokenizer.word_index) + 1

model = Sequential()
model.add(Embedding(input_dim=vocab_size, output_dim=16, input_length=max_len))
model.add(LSTM(16))
model.add(Dense(1, activation='sigmoid'))

# Compile model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Step 4: Train model
model.fit(padded_sequences, labels, epochs=10, verbose=1)

# Step 5: Predictions
predictions = model.predict(padded_sequences)

print("\nPredictions:")
print(predictions)
```

---

### Sample Output (Varies due to randomness)

```text id="lstm_out_q8"
Word Index:
{'this': 1, 'is': 2, 'i': 3, 'love': 4, 'project': 5, 'an': 6, 'amazing': 7, 'experience': 8,
'hate': 9, 'waiting': 10, 'in': 11, 'line': 12, 'the': 13, 'worst': 14, 'service': 15, 'absolutely': 16,
'fantastic': 17}

Sequences:
[[3, 4, 1, 5],
 [1, 2, 6, 7, 8],
 [3, 9, 10, 11, 12],
 [1, 2, 13, 14, 15],
 [16, 17]]

Padded Sequences:
[[ 3  4  1  5  0]
 [ 1  2  6  7  8]
 [ 3  9 10 11 12]
 [ 1  2 13 14 15]
 [16 17  0  0  0]]

Predictions:
[[0.81]
 [0.88]
 [0.12]
 [0.15]
 [0.79]]
```

# Question 9: Using spaCy, build a simple NLP pipeline (tokenization, lemmatization, and entity recognition)

**Answer:**

### Python Code

```python id="spacy_q9_pipeline"
import spacy

# Load spaCy English model
nlp = spacy.load("en_core_web_sm")

# Input text
text = """
Homi Jehangir Bhaba was an Indian nuclear physicist who played a key role in the development
of India’s atomic energy program. He was the founding director of the Tata Institute of Fundamental
Research (TIFR) and was instrumental in establishing the Atomic Energy Commission of India.
"""

# Process text through spaCy pipeline
doc = nlp(text)

# Tokenization and Lemmatization
print("TOKENS AND LEMMAS:")
for token in doc:
    if token.is_alpha:  # ignore punctuation
        print(f"{token.text:15} --> {token.lemma_}")

# Named Entity Recognition
print("\nNAMED ENTITIES:")
for ent in doc.ents:
    print(f"{ent.text:40} --> {ent.label_}")
```

---

### Sample Output

```text id="spacy_out_q9"
TOKENS AND LEMMAS:
Homi            --> Homi
Jehangir        --> Jehangir
Bhaba           --> Bhaba
was             --> be
an              --> an
Indian          --> indian
nuclear         --> nuclear
physicist       --> physicist
who             --> who
played          --> play
a               --> a
key             --> key
role            --> role
in              --> in
the             --> the
development     --> development
of              --> of
India           --> India
atomic          --> atomic
energy          --> energy
program         --> program
He              --> he
was             --> be
the             --> the
founding        --> found
director        --> director
of              --> of
the             --> the
Tata            --> Tata
Institute       --> Institute
of              --> of
Fundamental     --> Fundamental
Research        --> Research
TIFR            --> TIFR
was             --> be
instrumental    --> instrumental
in              --> in
establishing    --> establish
the             --> the
Atomic          --> Atomic
Energy          --> Energy
Commission      --> Commission
of              --> of
India           --> India

NAMED ENTITIES:
Homi Jehangir Bhaba                 --> PERSON
Indian                             --> NORP
India                              --> GPE
the Tata Institute of Fundamental Research --> ORG
TIFR                               --> ORG
the Atomic Energy Commission of India --> ORG
India                              --> GPE
```

# Question 10: Chatbot for a mental health platform using LSTM/GRU + spaCy/Stanford NLP

**Answer:**

## 1. Approach Overview

To build a chatbot for a mental health platform, we combine:

* **spaCy / Stanford NLP** → for text preprocessing and linguistic understanding
* **LSTM / GRU models** → for understanding sequential context in user messages
* **Intent classification + response generation** → to produce appropriate replies

The system is designed to understand emotional tone, detect intent (e.g., stress, anxiety, greeting), and respond in a supportive way.

---

## 2. Architecture

### A. Input Layer

User message (raw text)

### B. NLP Preprocessing (spaCy / Stanford NLP)

* Tokenization
* Lemmatization

* Named Entity Recognition (optional for context)

### C. Feature Representation

### D. Deep Learning Model (LSTM / GRU)

* LSTM or GRU layer captures sequential meaning and emotional context

### E. Response Generation

* Map predicted intent → predefined response template
* Optionally enhance using rule-based emotional tone adjustment

---

## 3. Data Preprocessing Pipeline

1. Convert text to lowercase
2. Tokenize using spaCy
3. Lemmatize tokens
4. Remove stopwords and punctuation
5. Convert tokens → sequences
6. Pad sequences to fixed length

---

## 4. Python Example (LSTM-based Intent Classifier Chatbot)

```python id="mental_chatbot_lstm"
import spacy
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense

# Load spaCy model
nlp = spacy.load("en_core_web_sm")

# Sample training data (intent classification)
texts = [
    "I am feeling very sad and anxious",
    "I am stressed about exams",
    "I feel happy today",
    "Hello how are you",
    "I need help I feel depressed",
    "Good morning"
]

labels = np.array([1, 1, 2, 3, 1, 3])  # 1=negative emotion, 2=positive, 3=neutral/greeting

# --------------------------
# Preprocessing with spaCy
# --------------------------
def preprocess(text):
    doc = nlp(text.lower())
    tokens = [token.lemma_ for token in doc if token.is_alpha and not token.is_stop]
    return " ".join(tokens)

processed_texts = [preprocess(t) for t in texts]

# Tokenization
tokenizer = Tokenizer()
tokenizer.fit_on_texts(processed_texts)
sequences = tokenizer.texts_to_sequences(processed_texts)

max_len = max(len(seq) for seq in sequences)
X = pad_sequences(sequences, maxlen=max_len, padding='post')

# --------------------------
# LSTM Model
# --------------------------
model = Sequential()
model.add(Embedding(input_dim=len(tokenizer.word_index)+1, output_dim=16))
model.add(LSTM(16))
model.add(Dense(3, activation='softmax'))

model.compile(loss='sparse_categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

model.fit(X, labels, epochs=10, verbose=1)

# --------------------------
# Simple chatbot response mapping
# --------------------------
responses = {
    1: "I'm really sorry you're feeling this way. I'm here for you.",
    2: "That's wonderful! I'm glad you're feeling positive.",
    3: "Hello! How can I assist you today?"
}

def chatbot_reply(text):
    clean = preprocess(text)
    seq = tokenizer.texts_to_sequences([clean])
    padded = pad_sequences(seq, maxlen=max_len, padding='post')
    pred = np.argmax(model.predict(padded), axis=1)[0]
    return responses.get(pred, "I'm here to listen.")

# Example interaction
print(chatbot_reply("I feel very stressed and anxious"))
```

---

## 5. Sample Output

```text id="chatbot_out_10"
Epoch 1/10 ...
Epoch 10/10 ...

"I'm really sorry you're feeling this way. I'm here for you."
```
